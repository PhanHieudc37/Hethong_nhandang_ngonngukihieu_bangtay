```python
video_path = "15.mp4"  # Tên video, để cùng thư mục thì chỉ cần tên

cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print(f"Không thể mở video: {video_path}")
else:
    while True:
        ret, frame = cap.read()  # Đọc frame

        if not ret:  # Kiểm tra xem đã đọc hết video chưa
            break

        # Giảm nhiễu
        frame_denoised = cv2.fastNlMeansDenoisingColored(frame, None, 10, 10, 7, 21)

        cv2.imshow("Video", frame_denoised)  # Hiển thị frame đã giảm nhiễu

        if cv2.waitKey(25) & 0xFF == ord('q'):  # Nhấn 'q' để thoát
            break

    cap.release()  # Giải phóng tài nguyên
    cv2.destroyAllWindows()
```
